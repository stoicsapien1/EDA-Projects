```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import seaborn as sns
```

```python
data={"DATA":[294.2,308.5,313.1,317.7,322.7,338.7]}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

```python
dmean=df["DATA"].mean()
dstd=df["DATA"].std()
df["Zj"]=(df["DATA"]-dmean)/dstd
df["uj"]=stats.norm.cdf(df["Zj"])
```

```python
df
```

```python
df["lnuj"]=np.log(df["uj"])
```

```python
df
```

```python
df["Zn+1-j"]=np.flip(df["Zj"].values)
```

```python
df
```

```python
df["U n+1-j"]=stats.norm.cdf(df["Zn+1-j"])
```

```python
df
```

```python
df["ln(1-Un+1-j)"]=np.log(1-df["U n+1-j"])
```

```python
df
```

```python
df["lnuj+lnUn+1-j"]=df["ln(1-Un+1-j)"]+df["lnuj"]
```

```python
df
```

```python
df["j"]=np.arange(1,len(df["DATA"])+1)
```

```python
df
```

```python
new_order=["j","DATA","Zj","uj","lnuj","Zn+1-j","U n+1-j","ln(1-Un+1-j)","lnuj+lnUn+1-j"]
df=df.reindex(columns=new_order)
```

```python
df
```

```python
df["(2j-1)*lnuj*lnUn+1-j"]=(2*df["j"]-1)*df["lnuj+lnUn+1-j"]
```

```python
df
```

```python
asum=df["(2j-1)*lnuj*lnUn+1-j"].sum()
```

```python
A2_statistic=-len(df["DATA"])-(1/len(df["DATA"]))*asum
```

```python
A2_statistic
```

```python
A2_modified = A2_statistic*(1+0.75/6+2.25/36)
```

```python
A2_modified
```

```python
critical_value=0.752
#Level of significance is 5%
if A2_modified<critical_value:
    print("Data is from the normal distribution")
else:
    print("Data is not from the normal distribution")
```
